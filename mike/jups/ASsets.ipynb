```python
import pandas as pd
```

```python
df = pd.read_csv('CARD_HIST12345.csv')
```

```python
df = df[['STAT_NCODE', 'NAME', 'TAXPAYER_NCODE', 'REGNUM_NCODE', 'UK']]
```

```python
df_final = pd.DataFrame()
```

```python
df_excel = pd.read_excel('46.71/2018.xlsx')
df_excel.columns = df_excel.iloc[2]
df_excel = df_excel.drop([0, 1, 2], axis=0)

a = ['2018, Нематериальные активы, RUB', '2018, Основные средства , RUB', '2018, Доходные вложения в материальные ценности, RUB', '2018, Долгосрочные финансовые вложения, RUB', '2018, Отложенные налоговые активы, RUB', '2018, Прочие внеоборотные активы, RUB', '2018, Внеоборотные активы, RUB', '2018, Чистые активы, RUB', '2018, Запасы, RUB', '2018, Дебиторская задолженность, RUB', '2018, Краткосрочные финансовые вложения, RUB', '2018, Денежные средства и денежные эквиваленты, RUB', '2018, Прочие оборотные активы, RUB', '2018, Оборотные активы, RUB', '2018, Активы  всего, RUB', '2018, Собственные акции, выкупленные у акционеров, RUB', '2018, Переоценка внеоборотных активов, RUB', '2018, Доходы будущих периодов , RUB']
b = ['INTANGIBLE_ASSETS_ATM', 'FIXED_ASSETS_ATM', 'PROFITABLE_INVESTMENTS_ATM', 'LONGTERM_INVESTMENTS_ATM', 'DEFERRED_TAX_ASSETS_ATM', 'OTHER_NONCURRENT_ASSETS_ATM', 'NONCURRENT_ASSETS_ATM', 'NET_ASSETS_ATM', 'RESERVES_ATM', 'RECEIVABLES_ATM', 'SHORTTERM_INVESTMENTS_ATM', 'CASH_CASH_EQUIVALENTS_ATM', 'OTHER_CURRENT_ASSETS_ATM', 'CURRENT_ASSETS_ATM', 'TOTAL_ASSETS_ATM', 'TREASURY_SHARES_ATM', 'REVALUATION_NONCURRENT_ASSETS_ATM', 'REVENUE_FUTURE_PERIODS_ATM']
c = ['Наименование', 'Регистрационный номер', 'Код налогоплательщика', 'Код статистики']
d = ['NAME', 'REGNUM_NCODE', 'TAXPAYER_NCODE', 'STAT_NCODE']
# a += c
ac = a
ac += c
# b += d
df_excel = df_excel[ac]
dic = dict(zip(ac, b))
df_excel.rename(columns=dic, inplace=True)
df_excel['Регистрационный номер'] = df_excel['Регистрационный номер'].fillna('-1').astype(int)
df_excel['Код налогоплательщика'] = df_excel['Код налогоплательщика'].fillna('-1').astype(int)
df_excel['Код статистики'] = df_excel['Код статистики'].fillna('-1').astype(int)
df_excel[b] = df_excel[b].fillna('-1').astype(float)

df_excel = df_excel.merge(df, how='inner', left_on = ['Наименование', 'Регистрационный номер', 'Код налогоплательщика', 'Код статистики'], right_on = ['NAME', 'REGNUM_NCODE', 'TAXPAYER_NCODE', 'STAT_NCODE'])
df_excel.drop(['Наименование', 'NAME', 'Регистрационный номер', 'Код налогоплательщика', 'STAT_NCODE', 'NAME', 'TAXPAYER_NCODE', 'REGNUM_NCODE', 'Код статистики'], axis=1, inplace=True)
df_excel.rename(columns={'UK': 'CARD_UK'}, inplace=True)

df_excel['VALUE_DAY'] = '20181231' #CHECK
df_excel['AS_OF_DAY'] = '20191214 12:20:00'
df_excel['DELETED_FLAG'] = 'N'
df_excel['JOB_INSERT'] = -1
df_excel['JOB_UPDATE'] = -1

df_final = pd.concat([df_final, df_excel], ignore_index=True)
```
